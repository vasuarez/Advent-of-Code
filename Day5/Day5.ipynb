```python
def polymer(string):
    x = list(string)
    for count, letter in enumerate(x):
        if (count+1) >= len(x):
            break
        if letter.lower() == x[count+1].lower():  # check if same type
            if letter != x[count+1]:  # check if different polarity
                x[count] = ''
                x[count+1] = ''
    x = ''.join(x)
    if x == string: #no changes were made
        return x
    return polymer(x)
```

```python
with open('input.txt', 'r') as f:
    for line in f:
        s = line.strip()
```

```python
polymer(s)
```
